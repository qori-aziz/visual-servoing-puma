# Code to draw data

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
kp35 = pd.read_csv("0.035.csv", sep=',')
kp5 = pd.read_csv("0.05.csv", sep=',')
kp7 = pd.read_csv("0.07new.csv", sep=',')
```

```python
err35 = kp35["normErr"]
err5 = kp5["normErr"]
err7 = kp7["normErr"]
```

```python
iter = np.arange(0, 133, 1)
fig, ax = plt.subplots()
line1 = ax.plot(iter,err35[:133], "b")
line2 = ax.plot(iter,err5[:133], "r")
line3 = ax.plot(iter, err7[:133], "g")
line4 = ax.axhline(y=100, linestyle = "--")
ax.set_xlabel("Iterations")
ax.set_ylabel("Normalized Error")
ax.set_title("Normalized Error vs Iterations")
ax.legend(["Kp: 0.035", "Kp: 0.05", "Kp: 0.07"], )
```

```python
joint1 = kp35["speed1"]
joint2 = kp35["speed2"]
joint3 = kp35["speed3"]
joint4 = kp35["speed4"]
joint5 = kp35["speed5"]
joint6 = kp35["speed6"]

fig, ax = plt.subplots()
iter = np.arange(0, joint1.shape[0], 1)
line1 = ax.plot(iter, joint1)
line2 = ax.plot(iter, joint2)
line3 = ax.plot(iter, joint3)
line4 = ax.plot(iter, joint4)
line5 = ax.plot(iter, joint5)
line6 = ax.plot(iter, joint6)
ax.set_xlabel("Iterations")
ax.set_ylabel("Speed (rad/s)")
ax.set_title("Speed of Joints for Kp = 0.035")
ax.legend(["Joint 1", "Joint 2", "Joint 3", "Joint 4", "Joint 5", "Joint 6"], )
```

```python
# Creating Shape of rectangle
xy = kp35[["u1","v1","u4","v4"]]
x0 = kp35["u1"][0]
y0 = kp35["v1"][0]
w0 = kp35["u4"][0] - kp35["u1"][0]
h0 = kp35["v4"][0] - kp35["v1"][0]
fig, ax = plt.subplots()
# ax.plot([0, 10],[0, 10])

x1 = kp35["u1"][kp35.shape[0]-1]
y1 = kp35["v1"][kp35.shape[0]-1]
w1 = kp35["u4"][kp35.shape[0]-1] - kp35["u1"][kp35.shape[0]-1]
h1 = kp35["v4"][kp35.shape[0]-1] - kp35["v1"][kp35.shape[0]-1]

plt.xlim(0, 320)
plt.ylim(0, 200)
rect = plt.Rectangle((x0,y0), w0, h0, fill=False, color="r", lw =3)
rect2 = plt.Rectangle((x1,y1), w1, h1, fill=False, color="g", lw =3)
ax.add_patch(rect)
ax.add_patch(rect2)
line1 = ax.plot(kp35["u1"], kp35["v1"], color="b")
line2 = ax.plot(kp35["u2"], kp35["v2"], color="b")
line3 = ax.plot(kp35["u3"], kp35["v3"], color="b")
line4 = ax.plot(kp35["u4"], kp35["v4"], color="b")
ax.invert_xaxis()
ax.invert_yaxis()
ax.set_xlabel("U (pixel)")
ax.set_ylabel("V (pixel)")
ax.set_title("Movements of Bounding Box for Kp = 0.035")
ax.legend(["Initial", "Final"], )
plt.show()

```
